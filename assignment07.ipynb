#  Import packages: 

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Give data and define function

```python
num     = 1001
std     = 5 
def fun(x):
 f = np.abs(x) * np.sin(x)
 return f
n       = np.random.rand(num)
nn      = n - np.mean(n)
x       = np.linspace(-10,10,num)
y1      = fun(x)
y2      = y1 + nn * std
```

# Plot the noisy data (x,y1) 

```python
plt.plot(x, y1, 'b.')
plt.show()
```

# Plot the clean data (x,y2) 

```python
plt.plot( x, y2, 'k.')
plt.show()
```

# Compute the optimal set A of model parameters 

```python
xa=[]
ya=[]
for i in range(0,9):
    xa.append(x**i)
xa=np.array(xa)
xA=xa.T
ya=np.array(y1)
y=ya.reshape(num,1)
A1=np.linalg.inv((xA.T).dot(xA))
A2=(xA.T).dot(y)
A=(A1).dot(A2)
print(A)
```

# Define the polynomial function and plot the curves

```python
Y=(xA).dot(A)
plt.plot(x, Y, 'b.')
```

# Plot the error  

```python
r=y1-Y
rr=sum(r**2)    
plt.plot(x, rr, 'k.')
```
